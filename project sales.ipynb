```python
import pandas as pd
sales_df=pd.read_csv("D:/OneDrive/Desktop/sample_sales_data.csv")
sales_df
    
```

```python
sales_df=sales_df.drop_duplicates()
sales_df
```

```python
sales_df.isnull()
```

```python
sales_group_df=sales_df.groupby("Region").agg({"Sales":['sum'],"Profit":['sum'],"Quantity":['sum','max']}).sum()
sales_group_df
```

```python
top_selling=sales_df.groupby(['Date','Product'])["Sales"].sum()
top_selling
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.xlabel("products")
plt.ylabel("top selling")
sns.histplot(top_selling)
plt.show()
```
